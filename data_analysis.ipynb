```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
data = []
with open("data/distance_measurement/exp1_distance_Denis_Grankin_25.txt", "r") as my_file:
    for line in my_file:
        datas = line.split() 
        if float(datas[0]) > 0:
            data.append(datas)

data =  np.array(data, dtype=float)


threshold = 1
diff = np.empty(data.shape)

diff[0] = np.inf  # always retain the 1st element
diff[1:] = np.diff(data)

mask = diff > threshold

new_arr = data[mask]

mask = data[:,1] > 0.35
far = data[mask]
print(far.shape)
min_dt_dist = 1

close = data[~mask]
fig, ax = plt.subplots()
ax.plot(close[:,0],close[:,1])
plt.show()
# print(argrelextrema(data, np.less))
```
